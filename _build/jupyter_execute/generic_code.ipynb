```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generic_pdf(x, pdf_func, fig, ax, shade = True,x_cdf_cutoff = None, x_shade_interval = None, 
                set_axes_origin = True, title = None,show_points=False):

    #Shade interval must be a list of two numbers 
    y = pdf_func(x)

    if(show_points):
        ax.plot(x, y,"k-o")
    else:
        ax.plot(x, y,"k-")

    if(shade):
        if(x_cdf_cutoff != None and x_shade_interval == None):
            x_shade = x[x<x_cdf_cutoff]
            y_shade = pdf_func(x_shade)
            ax.fill_between(x_shade, y_shade,color = "0.8")
        elif(x_shade_interval != None): 
            x_shade = x[np.logical_and(x>x_shade_interval[0],x<x_shade_interval[1])]
            y_shade = pdf_func(x_shade)
            ax.fill_between(x_shade, y_shade,color = "0.8")
        else: 
            ax.fill_between(x, y,color = "0.8")


    ax.spines['bottom'].set_position('zero')
    
    if(set_axes_origin):
        ax.spines['left'].set_position('zero')
        


    ax.spines['left'].set_linewidth(2)
    ax.spines['bottom'].set_linewidth(2)


    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)


    ax.set_xlabel('$x$',math_fontfamily = 'cm',fontsize = 14,loc = 'right')
    ax.set_ylabel('$f(x)$',math_fontfamily = 'cm',fontsize = 14,rotation = 'horizontal',loc = 'top')

    ax.set_title(title, y = 1.1, fontsize = 16)

    return fig,ax
```

```python
def f(x):
    f = (1/np.sqrt(2*np.pi))*np.exp(-1*(x**2)/2)
    return f
fig,ax = plt.subplots()


a = -0.5
x = np.linspace(-5,5,100)
fig,ax = generic_pdf(x, f, fig, ax, shade = False,title = '$\mathcal{N}(\\mu,\\sigma^2$)',x_cdf_cutoff=a)
# ax.plot([a,a],[0,f(a)],'k')
ax.set_xticks([])
ax.set_yticks([])
ax.text(0,-0.05,'$\\mu$',math_fontfamily = 'cm',fontsize = '14')
ax.text(0.5,f(0),'$\\frac{1}{\\sqrt{2\\pi\\sigma^2}}$',math_fontfamily = 'cm',fontsize = '14')

# ax.text(a-0.1,-0.02,'$b$',math_fontfamily = 'cm',fontsize = '14')
plt.savefig('Normal.png')
```
